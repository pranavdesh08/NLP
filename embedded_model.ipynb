```python
target = np.load('/content/drive/My Drive/Colab Notebooks/Capstone_project/target_data.npz')
target = target['arr_0']
```

```python
model = Doc2Vec.load("/content/drive/My Drive/Colab Notebooks/Capstone_project/embedded_vec.model")
```

```python
model.docvecs[1194696]
```

```python
vectors = []

for i in range(1194697):
    
    vectors.append(model.docvecs[i])
    
print(len(vectors))
print(vectors[:2])
```

```python
# Undersampling object
rus = RandomUnderSampler(sampling_strategy={1:26000,2:26000,3:26000,4:26000,5:26000},random_state=0)
def under_sample(x,y):
    
    X_resampled, y_resampled = rus.fit_resample(x,y)
    return X_resampled, y_resampled
```

```python
# training data
X_train, y_train = under_sample(vectors[0:1189697], target[0:1189697])
```

```python
X_test, y_test = vectors[1189697:] , target[1189697:]
```

```python
#X_train, X_test, y_train, y_test = train_test_split(vectors, target, test_size=0.33, random_state=42)
```

```python
clf_rf = RandomForestClassifier(random_state=0)

clf_rf = clf_rf.fit(X_train,y_train)

pred_rf = clf_rf.predict(X_test)

print(classification_report(y_test, pred_rf))

print('parameters used:',clf_rf.get_params())
```

```python
clf_rf = RandomForestClassifier(random_state=0, class_weight={1: 0.0001, 2: 0.4999, 3: 0.2999, 4: 0.2, 5: 0.0001})

clf_rf = clf_rf.fit(X_train,y_train)

pred_rf = clf_rf.predict(X_test)

print(classification_report(y_test, pred_rf))

print('parameters used:',clf_rf.get_params())
```

# WORD2VEC

```python
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
from imblearn.under_sampling import RandomUnderSampler #undersampling
```

```python
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Capstone_project/cleaned_reviews.csv')

df.head()
```

```python
word_idx = {}

f = open('/content/drive/My Drive/Colab Notebooks/Capstone_project/word_vectors.txt', encoding = 'utf-8')
next(f)
for line in f:
  
  values = line.split()
  word = values[0]
  coefs = np.asarray(values[1:])
  word_idx[word]=coefs
  
  
f.close()
```

```python
reviews = df['cleaned_reviews'].values.tolist()
tokens = np.zeros((len(reviews),55))
i=0
for ind,review in enumerate(reviews):
  
  token = word_tokenize(review)
  token = token[:55]
  for idx, wrd in enumerate(token): 
    if word_idx.get(wrd) is not None:
      tokens[ind,idx] = np.mean(np.array(word_idx.get(wrd)).astype(np.float)) 
  
  
  
  

print('total number of reviews:',len(tokens))
print(tokens[:5])
```

```python
# training data
X_train, y_train = under_sample(tokens[0:1189697], target[0:1189697])

X_test, y_test = tokens[1189697:] , target[1189697:]
```

```python
clf_rf = RandomForestClassifier(random_state=0)

clf_rf = clf_rf.fit(X_train,y_train)

pred_rf = clf_rf.predict(X_test)

print(classification_report(y_test, pred_rf))

print('parameters used:',clf_rf.get_params())
```
